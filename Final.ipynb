```python
Correlation_Matrix = df[["Transaction_Volume","Frequency_of_Transactions","Time_Since_Last_Transaction"]].corr()
Correlation_Matrix
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

"""Heat-Map of Corelation_Matrix"""

plt.figure(figsize = (10,6))
sns.heatmap(Correlation_Matrix, annot = True, cmap = "vlag")
plt.title("Correlation_Matrix")
plt.tight_layout()
plt.show()
```

```python
Top_Accounts = df.groupby("Account_Type")["Income"].sum().nlargest(5)
Top_Accounts
```

```python
"""Bar-Chart"""

Top_Accounts.plot(kind = "bar" , color = ["pink","lightblue"], width = 0.2)
plt.title("Top-5 Account by Income.")
plt.xlabel("Account_Type")
plt.ylabel("Income")
plt.tight_layout()
plt.show()
```

```python
Weekly_Avg_trans = df.groupby("Day_of_Week")["Transaction_Amount"].sum()

Weekly_Avg_trans
```

```python
"""Line-Chart"""

plt.figure(figsize = (6,6))
Weekly_Avg_trans.plot(kind = "line", marker = "o", color = "black", markerfacecolor = "pink", markersize = 6)
plt.title("Average Transaction Weekly.")
plt.xlabel("Day Of Week")
plt.ylabel("Average Transaction Amount")
plt.tight_layout()
plt.show()
```

```python
"""Scatter_Chart"""

plt.Figure(figsize = (10,10))

plt.scatter(df["Time_Since_Last_Transaction"],
            df["Frequency_of_Transactions"])

plt.xlabel("Time Since Last Transaction")
plt.ylabel("Frequency of Transactions")
plt.title("Time Since Last Transaction vs Frequency of Transactions")
plt.show()
```
