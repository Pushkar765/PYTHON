```python
df = pd.DataFrame({
    'Size' : ['M','L','S','L'],
    'Item' : ['Shirt','Pants','Shirt','Cap']
})
```

```python
df
```

```python
df['Size'] = pd.Categorical(df['Size'],categories = ['S','M','L'],ordered = True)
```

```python
df
```

```python
df.sort_values(by = 'Size')
```

```python
df
```
