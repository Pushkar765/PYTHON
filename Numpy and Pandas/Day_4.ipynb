```python
df2 = pd.DataFrame(
    [
        ["Pujen","Naitik","Aaryan"],
        [20,54,63],
        [99,97,93]
    ]
)

df2
```

## iloc & loc

```python
dd = [
    ["Alice",40,"Los-Angeles"],
    ["Linda",50,"Birmingham"],
    ["Jean",30,"Rome"],
]

df2 = pd.DataFrame(dd,columns=["Name","age","city"])

df2
```

```python
df2["Name"]
```

```python
df2.Name
```

```python
df2.Name[0]
```

```python
df2
```

```python
df2.iloc[0]
```

```python
df2.iloc[0:2]
```

```python
df2.iloc[0::2]
```

```python
df2.iloc[0:2,0:2]
```

```python
df2
```

```python
df2.loc[0]
```

#### How to set index from existing column.

```python
df3 = df2.set_index("Name")
```

```python
df3
```

```python
df3.loc["Linda"]
```

```python
df3.loc[["Alice","Jean"]]
```

```python
df3.loc["Alice":"Linda"]
```

```python
df3.loc[df3["city"] != "Birmingham"]
```
