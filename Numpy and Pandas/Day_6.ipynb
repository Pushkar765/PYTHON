### Merge and Join Data

```python
#First dataframe

df1 = pd.DataFrame({
        'Id' : [1,2,3,4],
        'Name' : ['Alice','Bob','Charlie','Daviid'],
        'Age' : [25,61,34,65]
})

#Second dataframe

df2 = pd.DataFrame({
      'Id' : [3,4,5,6],
      'Salary' : [12545,25156,58489,95147]  
})
```

```python
df1
```

```python
df2
```

#### Merge

```python
pd.merge(df1 , df2 , on = 'Id' , how = 'inner')
```

```python
pd.merge(df1 , df2 , on = 'Id' , how = 'left')
```

```python
pd.merge(df1 , df2 , on = 'Id' , how = 'outer')
```

```python
pd.merge(df1,df2)
```

```python
df1.merge(df2)
```

### Visualize Data

```python
data = {
    'Name' : ['Alice','Bob','Charlie','David','Emma'],
    'Age' : [25,41,62,15,54],
    'Salary' : [52361,21542,15423,15987,35789]
}

df = pd.DataFrame(data)

df
```

```python
import matplotlib 

df.plot(kind = "bar" , x = "Name" , y = "Salary" , title = "Salary by name" , xlabel = "Name" , ylabel = "Salary")
```

```python
df.plot(kind = "line" , x = "Name" , y = "Salary" , title = "Salary by name" , xlabel = "Name" , ylabel = "Salary" , marker = "o")
```

```python
df.plot(kind = "pie" , y = "Salary" , title = "Salary by Name" , labels = df["Name"] , startangle = 90 , autopct = '%0.2f%%')
```
