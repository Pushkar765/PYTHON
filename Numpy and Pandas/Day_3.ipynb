## Brodcasting of Array

#### Brodcasting is a powerful mechanism in numpy that allows Array Operations between arrays of different shapes without  copying data.

##### There are four types of array:
###### 1. Scalar
###### 2. Vector
###### 3. Matrix

```python
arr
```

#### 1) Scalar

```python
s1 = [3]
s2 = 5
s3 = np.array([6])
```

```python
arr + s2
```

```python
arr + s1
```

```python
arr + 9
```

```python
arr + s3
```

#### 2) Vector Brodcasting

##### (a) Row vector brodcasting(1D)

```python
a1 = np.array([[1,2,3],
               [4,5,6]])
a2 = np.array([10,20,30])

print(a1 + a2)
```

##### (b) Column vector brodcasting

```python
a1 = np.array([[1,2,3],
               [4,5,6]])
a2 = np.array([10,20]).reshape(-1,1)

print(a1 + a2)
```

#### 3) Matrix Brodcasting

```python
a1 = np.array([[1],[2]])
a2 = np.array([[10,20,30]])

print(a1)
print()
print(a2)
print()

a3 = a1 + a2 
print(a3)
```

```python
a1 = np.array([[1,2],[3,4]])
a2 = np.array([1,2,3])

print(a1)
print()
print(a2)

a3 = a1 + a2
print(a3)
```
