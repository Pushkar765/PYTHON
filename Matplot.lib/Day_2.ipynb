```python
import matplotlib.pyplot as plt
```

### Stack-Chart

```python
days = ["Mon","Tue","Wed","Thu","fri"]
product_A = [5,4,3,2,1]
product_B = [1,2,3,4,5]
product_C = [6,7,8,9,10]
```

```python
plt.bar(days,product_A,label = "Product_A")
plt.bar(days,product_B,label = "Product_B")
plt.bar(days,product_C,label = "Product_C")

plt.legend()
plt.show()
```

```python
plt.plot(days,product_A,label = "Product_A")
plt.plot(days,product_B,label = "Product_B")
plt.plot(days,product_C,label = "Product_C")

plt.legend()
plt.show()
```

```python
plt.bar(days,product_A,label = "Product_A",width = 0.5)
plt.bar(days,product_B,label = "Product_B",width = 0.3)
plt.bar(days,product_C,label = "Product_C",width = 0.1)

plt.legend(loc = "best")
plt.show()
```

```python
#Horizontal Stacked Bar-Chart

categories = ['A','B','C']
group_1 = [9,4,6]
group_2 = [6,2,5]

plt.barh(categories,group_1,label = 'Group-1')
plt.barh(categories,group_2,label = 'Group-2')

plt.title('Horizontal Stacked Bar')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.legend()
plt.tight_layout()
plt.show()
```
