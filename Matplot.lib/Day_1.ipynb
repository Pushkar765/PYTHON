```python
import matplotlib.pyplot as plt
```

### Bar Chart

```python
x = ['Math','Science','Gujarati','English']
y = [85,80,90,90]
```

```python
plt.bar(x,y)

plt.title("Subject-Wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x,y,width = 0.3)

plt.title("Subject-Wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x,y, color = ['red','green','blue','yellow'])

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x,y, color = ['blue','red','yellow','green'],width = 0.4)

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize = (4,2)) #figsize is basically use for sizing to image

plt.bar(x,y, align = 'center') #align is use to fix align of graphs

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x,y, color = "red", edgecolor = "blue" ,width = 0.5 , linewidth = 2)

plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

### Line chart

```python
months = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']

revenue = [12,51,42,36,1,5,65,84,97,56,45,99]
```

```python
plt.plot(months,revenue)
# use this if using a dataframe => plt.plot(df['months'], df['revenue'])

plt.title("Monthly Revenue")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.plot(months,revenue,marker = "o")
plt.show()
```

```python
plt.plot(months,revenue,marker = "+")
plt.show()
```

```python
plt.plot(months,revenue,marker = "*")
plt.show()
```

```python
plt.plot(months,revenue,marker = "o",markersize = 10, color = "black" , markerfacecolor = "red")
plt.show()
```

```python
plt.figure(figsize = (6,5))

plt.plot(months,revenue,marker = "o" , markersize = "5" , color = "black",markerfacecolor = "red" , linewidth = 2)
plt.title("Monthly_Revenue")
plt.xlabel("Months")
plt.ylabel("Revenue")

for i,v in enumerate(revenue):
    plt.text(i,v + 2,str(v),ha = "center", fontweight = "bold")

plt.grid(True,linestyle = "--",alpha = 0.7,axis = "both")
plt.savefig("Monthly_Revenue.png",dpi = 300)

plt.tight_layout()
plt.show()
```

### Scatter-Chart

```python
ad_spend = [1,2,3,4,4,3,2,1,5,6,7,9]
sales = [40,50,45,60,35,30,40,55,59,63,12,65]
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
```

```python
plt.scatter(ad_spend,sales)
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.scatter(ad_spend,sales,color = "green",edgecolor = "black",s=100)
#s is asize of marker

plt.xlabel("Ad-Spend")
plt.ylabel("Sales")

for i,month in enumerate(months):
    plt.text(ad_spend[i],sales[i] + 1.4,months[i],ha = "center",fontweight = "bold")

plt.tight_layout()
plt.show()
```

### Pie-Chart

```python
products = ["Laptops","Smart-Phones","Tablets"]
sales = [8,12,5]
```

```python
plt.pie(sales)
plt.show()
```

```python
plt.pie(sales,startangle = 90)
plt.show()
```

```python
plt.pie(sales,startangle = 90, labels = products, autopct = "%.2f%%")
plt.show()
```

```python
plt.pie(sales,startangle = 90, labels = products, autopct = "%.2f%%",colors=["Red","Pink","purple"])
plt.show()
```

```python
plt.pie(sales,startangle = 90, labels = products, autopct = "%.2f%%",explode = (0.1,0.2,0.08))
plt.show()
```

```python
plt.pie(
    sales,
    startangle = 90,
    labels = products,
    autopct = "%.2f%%",
    explode = (0.02,0.05,0.08),
    shadow = True,
    textprops = {"fontsize" : 14,"color" : "black"}
)

plt.show()
```

```python
plt.pie(sales,startangle = 90,labels = products,autopct = "%.2f%%",colors = ["tomato","pink","lightblue"])
centre_circle = plt.Circle((0,0),0.4,color = "white")
plt.gcf().gca().add_artist(centre_circle)

plt.show()
```

```python
plt.pie(
    sales,
    startangle = 90,
    labels=products,
    autopct=lambda pt: f'{pt:.2f}%'if pt > 20 else '',
    colors=["red","pink","blue"]
)

plt.legend(loc="best")
plt.text(0,0, "Sales", ha = "center" , va = "center", fontsize = 16)
plt.show()
```

### Histogram

```python
#sample data : Math exam score
scores = [45,50,50,62,22,36,12,36,49,78,
          20,51,46,64,36,16,25,34,43,61,
          25,36,54,74,75,76,71,72,73,71]
```

```python
plt.hist(scores)
```

```python
plt.hist(scores,bins = 20)
```

```python
plt.hist(scores,bins = [45,50,50,62],edgecolor = "black")
```

```python
#Compare two scores 

math_scores = [45,50,50,62,22,36,12,36,49,78,
              20,51,46,64,36,16,25,34,43,61,
              25,36,54,74,75,76,71,72,73,71]

science_score = [40,50,23,30,45,60,31,36,12,62,
                41,52,63,36,12,51,36,54,51,46,
                15,45,63,52,4,16,85,74,75,68]
```

```python
plt.hist(math_scores,bins = 5,alpha = 0.3,label = "Maths",color = "red",edgecolor = "black")
plt.hist(science_score,bins = 5,alpha = 0.3,label = "Science",color = "yellow",edgecolor = "black")
plt.legend(loc = "best")
plt.title("Comparison of Maths and Science scores.")
plt.show()
```
